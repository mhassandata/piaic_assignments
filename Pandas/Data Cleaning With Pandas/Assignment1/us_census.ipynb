```python
all_files = glob.glob('F:/PIAIC/Assignments/piaic_assignments/Pandas/Data Cleaning With Pandas/Assignment1/*.csv')
```

```python
dfs = []
for filename in all_files:
    df = pd.read_csv(filename, encoding='utf-8')
    dfs.append(df)
```

```python
dfs
```

```python
us_consus = pd.concat(dfs, ignore_index=True)
us_consus
```

```python
us_consus.columns
```

```python
us_consus.dtypes
```

```python
us_consus.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
us_consus.head()
```

```python
us_consus.info()
```

```python
# #Now convert the Income column into int from object
us_consus['Income'] = us_consus['Income'].str.replace(',', '')  # Remove comma from the income column
```

```python
us_consus['Income'] = us_consus['Income'].str.replace('$','')
```

```python
us_consus.head()
```

```python
us_consus['Income'] = pd.to_numeric(us_consus['Income']).astypepe(int)
```

```python
us_consus.info()
```

```python
#Now we will convert the "GenderPop" into two columns Men and Women
# Assuming that the GenderPop column is named 'GenderPop' in the DataFrame
us_consus[['Men', 'Women']] = us_consus['GenderPop'].str.split('_', expand=True)  # Split GenderPop column into Men and Women columns
```

```python
us_consus.head()
```

```python
us_consus['Men'] = us_consus['Men'].str.removesuffix('M')
us_consus['Women'] = us_consus['Women'].str.removesuffix('F')
us_consus.head()
```

```python
us_consus.info()
```

```python
us_consus['Men'] = pd.to_numeric(us_consus['Men']).astype(int)
```

```python
us_consus.info()
```

```python
us_consus['Women'] = pd.to_numeric(us_consus['Women'], errors='coerce').fillna(0).astype(int)
```
